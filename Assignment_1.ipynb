5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

1. Boxplot

```python
#SL vs Classification
sns.boxplot(x='SL',y='Classification',data=df)
plt.title('Sepal Length vs Classification')
plt.show()
```

```python
#SW vs Classification
sns.boxplot(x='SW',y='Classification',data=df)
plt.title('Sepal Width vs Classification')
plt.show()
```

```python
#PL vs Classification
sns.boxplot(x='PL',y='Classification',data=df)
plt.title('Petal Length vs Classification')
plt.show()
```

```python
#PW vs Classification
sns.boxplot(x='PW',y='Classification',data=df)
plt.title('Petal Width vs Classification')
plt.show()
```

2. Violin plot

```python
#SL vs Classification
sns.violinplot(x='Classification', y='SL', data=df)
plt.title('Sepal Length by Classification')
plt.show()
```

```python
#SW vs Classification
sns.violinplot(x='Classification', y='SW', data=df)
plt.title('Sepal Width by Classification')
plt.show()
```

```python
#PL vs Classification
sns.violinplot(x='Classification', y='PL', data=df)
plt.title('Petal Length by Classification')
plt.show()
```

```python
#PW vs Classification
sns.violinplot(x='Classification', y='PW', data=df)
plt.title('Petal Width by Classification')
plt.show()
```

3. Barplot

```python
#SL vs Classification
sns.barplot(x='Classification',y='SL',data=df)
plt.title('Sepal Length vs Classification')
plt.show()
```

```python
#SW vs Classification
sns.barplot(x='Classification',y='SW',data=df)
plt.title('Sepal Width vs Classification')
plt.show()
```

```python
#PL vs Classification
sns.barplot(x='Classification',y='PL',data=df)
plt.title('Petal Length vs Classification')
plt.show()
```

```python
#PW vs Classification
sns.barplot(x='Classification',y='PW',data=df)
plt.title('Petal Width vs Classification')
plt.show()
```

4. Scatter plot

```python
sns.scatterplot(data=df, x="SL", y="SW", hue="Classification")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.scatterplot(data=df, x="PL", y="PW", hue="Classification")
plt.title("Petal Length vs Petal Width")
plt.show()
```
